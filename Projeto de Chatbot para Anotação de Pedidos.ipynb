### Descrição do Projeto:

Imagine que você está criando um chatbot para a Pizzaria que Delícia, onde o funcionamento da pizzaria é de 17h às 3h da manhã. Os clientes podem fazer pedidos de pizzas disponíveis. O chatbot interagirá com os clientes, perguntando o sabor da pizza desejada e confirmará o preço, caso a pizza escolhida esteja no menu. O chatbot também será capaz de lidar com pedidos de múltiplas pizzas, calcular o valor total do pedido e fornecer uma lista final de pizzas pedidas e o valor total.

````python
## REGISTRO DE INFORMAÇÕES PADRÃO ##
## Aqui o dono da pizzaria registra os sabores disponíveis e valores de cada sabor (não atualiza o menu automaticamente)

#Sabores
sabores = ["CALABRESA", "QUEIJO","PEPPERONI","MARGHERITA","VEGETARIANA"]

#Preços
p_calab = 70.00
p_quei= 50.00
p_peppe = 80.00
p_marg = 80.00
p_veg = 80.00

#Listas
lista_pedido = []
lista_pagar = []

#INÍCIO DO BOT

## OPERAÇÃO APÓS RECEBER MENSAGEM DO CLIENTE ##

import random
hora = random.randint(1,24) # horário de funcionamento entre 17 às 03h
print("Início da conversa: " ,hora, "horas.")

if hora >= 17 or hora < 3: #Condição para que o restaurante esteja aberto e saudações por horário de início da conversa
  print("\n### ESTAMOS ABERTOS ###")

  if 18 > hora >= 17: # Quando a hora for menor que 18 - Boa tarde
      print("\nBoa tarde! - Seja bem vind@ ao ChatBot da Pizzaria que Delícia!\n\nSelecione uma das opções abaixo digitanto o texto informado entre parenteses para prosseguir: ")

  elif 23 >= hora >= 18: # Quando a hora estiver entre 18 e 23 (ambos inclusive) - Boa noite
      print("\nBoa noite! - Seja bem vind@ ao ChatBot da Pizzaria que Delícia!\n\nSelecione uma das opções abaixo digitanto o texto informado entre parenteses para prosseguir: ")

  elif hora > 23 or hora <= 3: # Quando a hora estiver entre 0 e 3 (ambos inclusive) - Bom dia
      print("\nBom dia! - Seja bem vind@ ao ChatBot da Pizzaria que Delícia!\n\nSelecione uma das opções abaixo digitanto o texto informado entre parenteses para prosseguir: ")

  print("\n############################")
  print("#  Visualizar menu (MENU)  #")
  print("#  Fazer pedido (PEDIDO)   #")
  print("############################")
  entrada_inicial = input("\nO que deseja fazer? ").upper() #Entrada para MENU ou PEDIDO

  while entrada_inicial != "MENU" and entrada_inicial != "PEDIDO": #Validação da entrada para MENU ou PEDIDO
    entrada_inicial = input("Ah, comando inválido! Digite MENU ou PEDIDO para seguir: ").upper()
  if entrada_inicial == "MENU" : #Selecionou MENU na opção inicial
      print("\nEsse é o nosso menu de pizzas!")
      print("\n| SABOR DA PIZZA  | PREÇO       |")
      print("| --------------- | ---------   |")
      print("| Calabresa       | R$ 70.00    |")
      print("| Queijo          | R$ 50.00    |")
      print("| Pepperoni       | R$ 80.00    |")
      print("| Margherita      | R$ 80.00    |")
      print("| Vegetariana     | R$ 80.00    |")

      saida_menu = input("\nGostaria de realizar seu pedido? (S/N): ").upper() #Realiza a pergunta para pedir ou acabar com a conversa!

  while saida_menu != "S" and saida_menu != "N": #Validação da resposta após ver o MENU (S/N)
      saida_menu = input("Ah, comando inválido! Gostaria de realizar seu pedido? (S/N): ").upper()
  if saida_menu == "N" : #Cliente desistiu da compra após ver o menu
    print("Ah, que pena! Quem sabe da próxima vez, hein!?")
    pass #Passa as partes de PEDIDO e PRECO
  pass #Passa a mensagem de FECHADO


## INICIANDO O PEDIDO ##

  if saida_menu == "S" or entrada_inicial == "PEDIDO": #Selecionou PEDIDO na opção inicial sem ver o menu
    print("\n#### Ótimo! Vamos iniciar seu pedido! ####")
    nome_cliente = input("Insira seu nome: ").title() #Title para arrumar o nome
    end_cliente = input("Insira o endereço para entrega: ").title()
    pagamento_cliente = input("Qual a forma de pagamento? (Dinheiro, Cartão ou Pix): ").title()

    pedido = input("\nDigite o sabor da pizza desejada: ").upper() #Input do sabor da pizza (repetição da entrada)

    if pedido == "CALABRESA": #Informa qual sabor foi adicionado ao pedido e o valor
      print("Adicionado Pizza Calabresa ao pedido - Valor R$ 70.00")
    elif pedido == "QUEIJO":
      print("Adicionado Pizza Queijo ao pedido - Valor R$ 50.00")
    elif pedido == "PEPPERONI":
      print("Adicionado Pizza Pepperoni ao pedido - Valor R$ 80.00")
    elif pedido == "MARGHERITA":
      print("Adicionado Pizza Margherita ao pedido - Valor R$ 80.00")
    elif pedido == "VEGETARIANA":
      print("Adicionado Pizza Vegetariana ao pedido - Valor R$ 80.00")
    while pedido != "FINALIZAR":
        if pedido not in sabores: #Validação do sabor solicitado (se há no menu)
              pedido = input("\nAh, esse sabor não existe! Tente novamente ou digite FINALIZAR para encerrar: ").upper() #Se não possuir no menu, TENTE NOVAMENTE
        else:
              lista_pedido.append(pedido) #Se possuir o sabor no MENU, entra na lista_pedido para posterior análise de preço
              pedido = input("\nGostaria de adicionar alguma outra pizza ao pedido? Se não digite FINALIZAR: ").upper() #Refaz a pergunta para continuar pedindo ou encerrar

        if pedido == "CALABRESA": #Informa qual sabor foi adicionado ao pedido e o valor
              print("Adicionado Pizza Calabresa ao pedido - Valor R$ 70.00")
        elif pedido == "QUEIJO":
              print("Adicionado Pizza Queijo ao pedido - Valor R$ 50.00")
        elif pedido == "PEPPERONI":
              print("Adicionado Pizza Pepperoni ao pedido - Valor R$ 80.00")
        elif pedido == "MARGHERITA":
              print("Adicionado Pizza Margherita ao pedido - Valor R$ 80.00")
        elif pedido == "VEGETARIANA":
              print("Adicionado Pizza Vegetariana ao pedido - Valor R$ 80.00")

## CÁLCULO DE PREÇO ##

    total_pagar = 0

    for i in lista_pedido: #Adiciona o valor de cada pizza conforme SABOR inserido na LISTA_PEDIDO e valor de cada sabor informado no registro (primeiras linhas)
      if i == "CALABRESA":
        total_pagar = total_pagar + p_calab
        lista_pagar.append(total_pagar)
        continue
      if i == "QUEIJO":
        total_pagar = total_pagar + p_quei
        lista_pagar.append(total_pagar)
        continue
      if i == "PEPPERONI":
        total_pagar = total_pagar + p_peppe
        lista_pagar.append(total_pagar)
        continue
      if i == "MARGHERITA":
        total_pagar = total_pagar + p_marg
        lista_pagar.append(total_pagar)
        continue
      if i == "VEGETARIANA":
        total_pagar = total_pagar + p_veg
        lista_pagar.append(total_pagar)
        continue

    print("\n##### SEU PEDIDO FOI CONFIRMADO! #####") #Mensagem de fechamento do pedido
    print(f"\nAgradeçemos a sua preferência, {nome_cliente}!")
    print(f"O pedido será entregue no endereço {end_cliente}!")
    print(f"A forma de pagamento escolhida foi {pagamento_cliente}!")
    print("\nSegue a lista das pizzas pedidas: ")

    contador = 1 #Contador para enumerar as pizzas inicianco em 1
    for elemento in range(len(lista_pedido)):
      print(contador,"-", lista_pedido[elemento])
      contador += 1

    print("\nO valor total da compra foi de: R$", round(total_pagar,2))

else:
  print("\nAh, infelizmente estamos fechados no momento!")
  print("\nEntre em contato conosco entre 17h e 03h e garanta a melhor pizza da cidade! \nAté mais!")
````
